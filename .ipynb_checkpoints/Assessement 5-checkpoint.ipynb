```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns 

from sklearn.model_selection import train_test_split
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
from sklearn.preprocessing import RobustScaler
from sklearn.metrics import balanced_accuracy_score
from sklearn.feature_selection import f_regression
from sklearn.feature_selection import mutual_info_regression
```

```python
train_df = pd.read_csv("ML-A5-2022_train.csv")
```

```python
ax = sns.countplot(train_df.label,label="Count")
Healthy, Tumor = train_df.label.value_counts()
print('Number of Tumor: ',Tumor)
print('Number of Healthy : ',Healthy)
```

```python
# Preprocessing 
```

```python
# Removing null values 
for i in train_df.columns[train_df.isnull().any(axis=0)] : 
    train_df[i].fillna(0, inplace=True)
    
# Transforming categorical features 
df = train_df.replace(["low", "medium", "high"], [0, 1, 2])
```

```python
df.head()
```

```python
lst = []
# Removing columns of 0 and full of same values 
for i in df.columns : 
    if(np.count_nonzero(train_df[i]) == 0) :
        lst.append(i)
        
df.drop(labels = lst, axis = 1, inplace = True)
```

```python
df.head()
```

```python
lst = []
for i in df.columns : 
    if(np.all(df[i] == df[i][0])) :
        lst.append(i)

df.drop(labels = lst, axis = 1, inplace = True)
```

```python
df = pd.read_csv("dataset.csv")
```

```python
df.head()
```

```python
for i in df.columns :
    if(i == "label") : continue
    df[i] = np.abs(df[i])
    df[i] /= np.max(df[i])
```

```python
#df.describe()
print(df.label)
```

```python
df_s = df.sample(frac = 1, random_state = 42)
test_size = 0.3
df_train = df_s.iloc[:700, :]
df_test = df_s.iloc[700:, :]
```

```python
from sklearn.svm import SVC, LinearSVC
from sklearn.linear_model import Perceptron
from sklearn.linear_model import SGDClassifier
```

```python
perceptron = Perceptron()
linear_svc = LinearSVC()
sgd = SGDClassifier()

classifiers = [perceptron, linear_svc, sgd]
```

```python
evaluate = []
models = ['Perceptron', 'Linear SVC', 'Stochastic Gradient Descent']

for i in range(len(models)) :
    clf = classifiers[i]
    clf.fit(df_train.iloc[:, :-1], df_train.iloc[:, -1])
    BCR = balanced_accuracy_score(df_test.iloc[:, -1], clf.predict(df_test.iloc[:, :-1]))
    evaluate.append(BCR)

c_bcr = pd.DataFrame({'Model' : models, 'BCR' : evaluate})
c_bcr.sort_values(by='BCR', ascending=False)
```

```python
clf = sgd
test_bcr = []

for i in range(100) :
    test = pd.DataFrame.sample(df_test, n = 100, random_state=i)
    BCR = balanced_accuracy_score(test.iloc[:, -1], clf.predict(test.iloc[:, :-1]))
    test_bcr.append(BCR)
    
mean_test_bcr = sum(test_bcr)/100
print(mean_test_bcr)
```

```python
frame = pd.DataFrame(columns = ["indiv_test_acc", "CI_lower_bound", "CI_upper_bound", 
                                "mean_test_acc", "observed_lower_bound", "observed_upper_bound"])

train = pd.DataFrame.sample(df_train,frac=1, random_state = 42)
clf = LinearSVC()

for i in range(50) : 

    test = pd.DataFrame.sample(df_test,n = 100, random_state=i)


    
    clf.fit(train.iloc[:, :-1], train.iloc[:, -1])

    test_bcr = balanced_accuracy_score(test.iloc[:, -1], clf.predict(test.iloc[:, :-1]))


    n = test.shape[0]
    zn = 1.96
    x = zn * np.sqrt((test_bcr*(1-test_bcr))/n)
    
    
    test_bcrs = []

    for j in range(100) :
        test = pd.DataFrame.sample(test,n = 100, random_state=(i+1)*j)
        
        test_bcrs.append(balanced_accuracy_score(test.iloc[:, -1], clf.predict(test.iloc[:, :-1])))
        
    mean_test_bcr = sum(test_bcrs)/100
    print(mean_test_bcr)

    frame = frame.append(pd.DataFrame(data = np.array([[test_bcr, 
                                                        test_bcr-x, test_bcr+x, 
                                                        mean_test_bcr, np.quantile(test_bcrs, 0.025), 
                                                        np.quantile(test_bcrs, 0.975)]]), 
                                      columns= ["indiv_test_acc", "CI_lower_bound", 
                                                "CI_upper_bound", "mean_test_acc", 
                                                "observed_lower_bound", "observed_upper_bound"]), 
                         ignore_index = True)
```
